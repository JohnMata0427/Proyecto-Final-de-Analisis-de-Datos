```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
data1 = pd.read_csv('Datos iniciales/noticia1.csv')
data1
```

```python
data1.duplicated().sum()
```

```python
data1.drop_duplicates(inplace=True)
```

```python
data1.isnull().sum()
```

```python
# Rellenar los valores nulos en columnas de texto con 'sin información'
data1.loc[:, data1.dtypes == object] = data1.loc[:, data1.dtypes == object].fillna('sin información')

# Rellenar los valores nulos en columnas numéricas con 0
data1.loc[:, data1.dtypes != object] = data1.loc[:, data1.dtypes != object].fillna(0)
```

```python
# Exportar a MySQL

engine = create_engine('mysql+pymysql://root:12345@localhost:3307/noticias_1')

data1.to_sql('noticias_1', con=engine, if_exists='replace', index=False)
```

```python
data1 = pd.read_sql('SELECT * FROM noticias_1', engine)
data1
```

```python
data1.to_csv('Datos limpios/noticias_1.csv', index=False)
```

```python
#SEGUNDA DATA SET
data2 = pd.read_excel('Datos iniciales/noticia2.xlsx')
data2
```

```python
data2.duplicated().sum()
```

```python
data2.isnull().sum()
```

```python
# Rellenar los valores nulos en columnas de texto con 'sin información'
data2.loc[:, data2.dtypes == object] = data2.loc[:, data2.dtypes == object].fillna('sin información')

# Rellenar los valores nulos en columnas numéricas con 0
data2.loc[:, data2.dtypes != object] = data2.loc[:, data2.dtypes != object].fillna(0)
```

```python
# Exportar a MongoDB

from pymongo import MongoClient

client = MongoClient('localhost', 27017)
db = client['noticias_2']
collection = db['noticias_2']

data2.reset_index(inplace=True)
data_dict = data2.to_dict("records")

collection.insert_many(data_dict)
```

```python
data2 = pd.DataFrame(list(collection.find()))

data2
```

```python
data2.to_json('Datos limpios/noticias_2.json', orient='records', default_handler=str)
```
