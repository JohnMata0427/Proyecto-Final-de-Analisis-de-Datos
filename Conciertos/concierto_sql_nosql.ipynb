```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
data1 = pd.read_csv('./Datos iniciales/concierto1.csv')
data1
```

```python
data1.duplicated().sum()
```

```python
data1.drop_duplicates(inplace=True)
```

```python
data1.isnull().sum()
```

```python
# Rellenar los valores nulos en columnas de texto con 'sin información'
data1.loc[:, data1.dtypes == object] = data1.loc[:, data1.dtypes == object].fillna('sin información')

# Rellenar los valores nulos en columnas numéricas con 0
data1.loc[:, data1.dtypes != object] = data1.loc[:, data1.dtypes != object].fillna(0)
```

```python
# SQLite

engine = create_engine('sqlite:///concierto_1.db')
data1.to_sql('concierto_1', engine, index=False)
```

```python
data1 = pd.read_sql('concierto_1', engine)
data1
```

```python
data1.to_csv('./Datos limpios/concierto1.csv', index=False)
```

```python
data4 = pd.read_csv('./Datos iniciales/concierto4.csv')
data4
```

```python
data4.duplicated().sum()
```

```python
data4.isnull().sum()
```

```python
# Rellenar los valores nulos en columnas de texto con 'sin información'
data4.loc[:, data4.dtypes == object] = data4.loc[:, data4.dtypes == object].fillna('sin información')

# Rellenar los valores nulos en columnas numéricas con 0
data4.loc[:, data4.dtypes != object] = data4.loc[:, data4.dtypes != object].fillna(0)
```

```python
from pymongo import MongoClient

client = MongoClient('localhost', 27017)

# Crear una base de datos llamada 'concierto_4'
db = client['concierto_4']

# Crear una colección llamada 'concierto_4'
collection = db['concierto_4']

# Enviar el DataFrame a MongoDB
data4.reset_index(inplace=True)

data_dict = data4.to_dict('records')

collection.insert_many(data_dict)
```

```python
data4 = pd.DataFrame(list(collection.find()))
data4
```

```python
data4.to_json('./Datos limpios/concierto4.json', orient='records', default_handler=str)
```
