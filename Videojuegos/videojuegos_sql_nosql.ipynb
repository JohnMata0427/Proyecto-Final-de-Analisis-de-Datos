```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
data1 = pd.read_csv('./Datos iniciales/videojuegos1.csv')
data1
```

```python
data1.duplicated().sum()
```

```python
data1.isnull().sum()
```

```python
# Rellenar los valores nulos en columnas de texto con 'sin información'
data1.loc[:, data1.dtypes == object] = data1.loc[:, data1.dtypes == object].fillna('sin información')

# Rellenar los valores nulos en columnas numéricas con 0
data1.loc[:, data1.dtypes != object] = data1.loc[:, data1.dtypes != object].fillna(0)
```

```python
data1.to_csv('./Datos limpios/videojuegos_1.csv', index=False)
```

```python
data4 = pd.read_csv('./Datos iniciales/videojuegos4.csv')
data4
```

```python
data4.duplicated().sum()
```

```python
data4.isnull().sum()
```

```python
# Rellenar los valores nulos en columnas de texto con 'sin información'
data4.loc[:, data4.dtypes == object] = data4.loc[:, data4.dtypes == object].fillna('sin información')

# Rellenar los valores nulos en columnas numéricas con 0
data4.loc[:, data4.dtypes != object] = data4.loc[:, data4.dtypes != object].fillna(0)
```

```python
import urllib
import pyodbc

# Conectar a SQL Server con autenticación de Windows
params = urllib.parse.quote_plus("DRIVER={ODBC Driver 17 for SQL Server};SERVER=KIYOTAKA;DATABASE=videojuegos_4;Trusted_Connection=yes;")
engine = create_engine("mssql+pyodbc:///?odbc_connect=%s" % params)

# Enviar el DataFrame a SQL Server
data4.to_sql('videojuegos_4', con=engine, if_exists='replace', index=False)
```

```python
data4 = pd.read_sql('SELECT * FROM videojuegos_4', engine)
data4
```

```python
data4.to_csv('./Datos limpios/videojuegos_4.csv', index=False)
```
