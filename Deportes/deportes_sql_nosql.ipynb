```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
data1 = pd.read_csv('./Datos iniciales/deportes1.csv')
data1
```

```python
data1.duplicated().sum() # Se comprueba si hay valores duplicados
```

```python
data1.isnull().sum() # Se comprueba si hay valores nulos
```

```python
# Rellenar los valores nulos en columnas de texto con 'sin información'
data1.loc[:, data1.dtypes == object] = data1.loc[:, data1.dtypes == object].fillna('Sin información')

# Rellenar los valores nulos en columnas numéricas con 0
data1.loc[:, data1.dtypes != object] = data1.loc[:, data1.dtypes != object].fillna(0)
```

```python
# Exportar el archivo limpio como JSON
data1.to_json('./Datos limpios/deportes_1.json', orient='records')
```
